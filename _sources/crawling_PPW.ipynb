```python
import requests
import pandas as pd
```

```python
api_key = "e77104b8635e01c147fff3980869bdad"  # ganti dengan API Key Springer kamu
keywords = ["web mining", "web usage mining"]

all_results = []

for keyword in keywords:
    url = "https://api.springernature.com/meta/v2/json"
    params = {
        "q": keyword,
        "api_key": api_key,
        "p": 10   # jumlah hasil per query
    }

    response = requests.get(url, params=params)

    if response.status_code == 200:
        data = response.json()
        total = data['result'][0]['total']
        print(f"Kata kunci: {keyword} | Total hasil: {total}")

        for record in data['records']:
            doi = record.get('doi', 'N/A')
            title = record.get('title', 'No title')
            abstract = record.get('abstract', 'No abstract')
            url_record = record.get('url', [{'value':'N/A'}])[0]['value']

            all_results.append({
                "keyword": keyword,
                "doi": doi,
                "title": title,
                "abstract": abstract,
                "url": url_record
            })
    else:
        print("Error:", response.status_code, response.text)
```

```python
# Simpan ke CSV
df = pd.DataFrame(all_results)
df.to_csv("crawling_springer.csv", index=False)
print("\nData berhasil disimpan ke crawling_springer.csv")
df.head()
```
